```python
from tensorflow.keras.datasets import cifar10
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
train_images = train_images.reshape((train_images.shape[0], train_images.shape[1] * train_images.shape[2] * train_images.shape[3])).astype("float32")/255
test_images = test_images.reshape((test_images.shape[0], test_images.shape[1] * test_images.shape[2] * test_images.shape[3])).astype("float32")/255
```

```python
split_val = 5000
train_images, val_images = train_images[split_val:], train_images[:split_val]
train_labels, val_labels = train_labels[split_val:], train_labels[:split_val]
```

```python
train_images
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels, dtype ="uint8")
test_labels = tf.keras.utils.to_categorical(test_labels, dtype ="uint8")
```

```python
class LogisticRegression(keras.Model):
    def train_step(self, data):
        if len(data) == 3:
            input, y, sample_weight = data
        else:
            sample_weight = None
            input, y = data

        with tf.GradientTape() as tape:
            y_pred = self(input, training=True)
            loss = self.compiled_loss(
                y,
                y_pred,
                sample_weight=sample_weight,
                regularization_losses=self.losses,
            )

        trainable_vars = self.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)

        self.optimizer.apply_gradients(zip(gradients, trainable_vars))

        self.compiled_metrics.update_state(y, y_pred, sample_weight=sample_weight)

        return {m.name: m.result() for m in self.metrics}
```

```python
train_is = train_images.shape
train_images[4]
val_labels
```

```python
inputs = keras.Input(shape=(32 * 32 * 3, ))
features = layers.Dense(1024, activation="relu")(inputs)
features = layers.Dropout(0.5)(features)
outputs = layers.Dense(10, activation="softmax")(features)
model = LogisticRegression(inputs, outputs)
```

```python
model.compile(
    optimizer=keras.optimizers.RMSprop(),
    loss=keras.losses.MeanAbsoluteError(),
    metrics=[keras.metrics.MeanAbsoluteError()])
```

```python
model.fit(
    train_images,
    train_labels,
    epochs=10,
)
```
